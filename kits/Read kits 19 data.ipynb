```
def visualise_slice(volume_data, segmentation_data, idx):
    fig, ax = plt.subplots(1, 2, figsize=(16, 8))
    plt.close()

    ax[0].set_title('Scan')
    ax[0].imshow(volume_data[idx], cmap='gray')
    ax[1].set_title('Segmentation')
    ax[1].imshow(segmentation_data[idx], cmap='hot')
    return fig
```
